```python
import numpy as np
import pandas as pd
```

```python
path="Data Set 1.csv"
```

```python
df=pd.read_csv(path)
```

```python
df
```

```python
df.drop(["Index"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["Age"].fillna((df["Age"].mean()),inplace=True)
```

```python
df.head()
```

```python
df["Age"]=df["Age"].astype(int)
```

```python
df.head()
```

```python
df["Location"]=df["Location"].str.replace(",In","").str.strip()
df["Location"]=df["Location"].str.replace(",Ny","").str.strip()
df["Location"]=df["Location"].str.replace("Aus","").str.strip()
```

```python
df
```

```python
df["Location"]=df["Location"].str.replace("tralia","Australia").str.strip()
df["Location"]=df["Location"].str.replace("India In","India").str.strip()
```

```python
df
```

```python
df["Salary"]=df["Salary"].str.replace("k","000").str.strip()
df["Salary"]=df["Salary"].str.replace("$","").str.strip()
```

```python
df.head()
```

```python
df[["Starting Salary","Ending Salary"]]=df["Salary"].str.split('-',expand=True)
df=df.drop(columns=["Salary"])
```

```python
df
```

```python
df.dtypes
```

```python
df["Starting Salary"]=df["Starting Salary"].astype(int)
df["Ending Salary"]=df["Ending Salary"].astype(int)
```

```python
df.dtypes
```

```python
df["Established"]=df["Established"].replace(-1,np.nan)
```

```python
df.head()
```

```python
df["Established"].fillna((df["Established"].mean()),inplace=True)
```

```python
df["Established"]=df["Established"].astype(int)
```

```python
df.head()
```

```python
df["Easy Apply"]=df["Easy Apply"].str.replace('-1',"FALSE")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["Rating"]=df["Rating"].replace(-1,0)
```

```python
df["Rating"].fillna((df["Rating"].mean()),inplace=True)
df["Rating"]=df["Rating"].round(1)
```

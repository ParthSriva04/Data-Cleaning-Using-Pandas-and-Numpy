```python
import pandas as pd 
import numpy as np
```

```python
path="Data Set 2.tsv"
```

```python
df=pd.read_csv(path,sep='\t')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["item_price"]=df["item_price"].str.replace('$',"")
```

```python
df["item_price"]=df["item_price"].astype(float)
```

```python
df.dtypes
```

```python
df["item_name"].unique
```

```python
df["choice_description"].fillna("unknown",inplace=True)
```

```python
df.head()
```

```python
df["choice_description"]=df["choice_description"].str.replace("[","").str.strip()
df["choice_description"]=df["choice_description"].str.replace("]","").str.strip()
```

```python
df.head()
```

```python
df.head()
```

```python
df["item_name"]=df["item_name"].str.replace('[^a-zA-Z0-9,\s]','',regex=True)
df["choice_description"]=df["choice_description"].str.replace('[^a-zA-Z0-9,\s]','',regex=True)
```

```python
df.reset_index(drop=True,inplace=True)
```
